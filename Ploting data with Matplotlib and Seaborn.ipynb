### Plotting graph For IRIS Dataset Using Seaborn Library And matplotlib.pyplot library

#### A. Creating and customizing line charts using Matplotlib 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.image as mpimg

df=pd.DataFrame({'x_values': range(1,11), 'y_values': np.random.randn(10) })

# Draw plot
plt.plot( 'x_values', 'y_values', data=df, color='skyblue')
plt.show()

# Draw line chart by modifiying transparency of the line
plt.plot( 'x_values', 'y_values', data=df, color='skyblue', alpha=0.3)

# Show plot
plt.show()
```

```python
# Draw line chart with dashed line
plt.plot( 'x_values', 'y_values', data=df, linestyle='dashed')

# Show graph
plt.show()

# Draw line chart with dashed line
plt.plot( 'x_values', 'y_values', data=df, linestyle='-.')

# Show graph
plt.show()

# Draw line chart with dashed line
plt.plot( 'x_values', 'y_values', data=df, linestyle='--')

# Show graph
plt.show()

# Draw line chart with dashed line
plt.plot( 'x_values', 'y_values', data=df, linestyle=':')

# Show graph
plt.show()
```

```python
# Modify line width of the graph
plt.plot( 'x_values', 'y_values', data=df, linewidth=22)

# Show graph
plt.show()
```

#### B. Visualizing relationships between two or more variables using scatter plots

```python
# Fixing random state for reproducibility
np.random.seed(19680801)


N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time");
```

```python
penguins = sns.load_dataset("penguins")
sns.pairplot(penguins);
```

```python
sns.pairplot(penguins, hue="species");
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species");
```

```python
x = np.random.randn(100)
y1 = np.random.randn(100)
plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
sns.jointplot(x=x,y=y1);
```

#### C. Studying distributions of variables using histograms & bar charts

```python
df = sns.load_dataset("iris")
sns.distplot(df['sepal_length'])
plt.show();
```

```python
sns.distplot(df['sepal_length'],bins=25);
```

```python
sns.distplot(a=df['sepal_length'],hist=True,kde=True,rug=True);
```

```python
sns.distplot(a=df['sepal_length'],rug=True,rug_kws={'color':'r','alpha':0.35,'linewidth':5});
```

```python
sns.distplot(df['sepal_length'],color='skyblue',label='Sepal length')
sns.distplot(df['sepal_width'],color='lightgreen',label='Sepal width')
plt.show()
```

```python
# Bar Plot

df = sns.load_dataset('titanic')
sns.barplot(x = 'class', y = 'fare', data = df)
plt.show()
```

```python
sns.barplot(x = 'class', y = 'fare', hue = 'sex', data = df)
plt.show()
```

```python
sns.barplot(x = 'class', y = 'fare', data = df, order = ["Third", "Second", "First"])
plt.show()
```

#### D. Visualizing two-dimensional data using heatmaps 

```python
data = np.random.random(( 12 , 12 ))
plt.imshow( data )
plt.title( "2-D Heat Map" )
plt.show()
```

```python
plt.imshow(data, interpolation='nearest')
plt.colorbar()
plt.title( "2-D Heat Map" )
plt.show()
```

```python
sns.heatmap(data=data, annot=True)
plt.show()
```

#### F. Plotting multiple Matplotlib and Seaborn charts in a grid

```python
df = sns.load_dataset('penguins')
sns.relplot(data=df, x='bill_length_mm', y='bill_depth_mm', col='island')
plt.show()
```

```python
sns.relplot(data=df, x='bill_length_mm', y='bill_depth_mm', row='sex', col='island')
plt.show()
```

```python
sns.relplot(data=df, x='bill_length_mm', y='bill_depth_mm', row='sex', col='island', hue='species')
plt.show()
```
